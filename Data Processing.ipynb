```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Street_Tree_List-2022-01-30_FILTERED.csv")
```

```python
planted_df = df[df['PlantDate'].notna()] # filtering to trees with planting date
```

```python
planted_df
```

```python
planted_df["formatted_date"] = pd.to_datetime(planted_df['PlantDate'], format='%m/%d/%y %H:%M')
planted_df['month_planted'] = planted_df['formatted_date'].dt.month
planted_df['year_planted'] = planted_df['formatted_date'].dt.year

# since datetime attributes a %y year below 69 a century of 2000, which can create erroneous years in the future instead of in the 1900s
year_correction = planted_df['year_planted'] > 2025 
planted_df.loc[year_correction, 'year_planted'] -= 100

planted_df["season"] = planted_df['month_planted'].apply(lambda x: 
                                                "spring" if x in [3, 4, 5] 
                                                else "summer" if x in [6, 7, 8] 
                                                else "fall" if x in [9, 10, 11]
                                                else "winter" if x in [12, 1, 2]
                                                else None)
```

```python
final_df = planted_df[['TreeID', 'Latitude', 'Longitude', 'month_planted', 'year_planted', 'season']].copy()
final_df.rename(columns={'TreeID': 'tree_id', 'Latitude': 'lat', 'Longitude': 'lon'}, inplace=True)
```

```python
final_df
```

```python
final_df.to_csv("SF_trees_final.csv")
```

```python
month_counts = final_df['month_planted'].value_counts()

month_counts_df = pd.DataFrame(month_counts.reset_index())
month_counts_df.columns = ['month', 'count']

month_counts_df["season"] = month_counts_df['month'].apply(lambda x: 
                                                "spring" if x in [3, 4, 5] 
                                                else "summer" if x in [6, 7, 8] 
                                                else "fall" if x in [9, 10, 11]
                                                else "winter" if x in [12, 1, 2]
                                                else None)
month_counts_df = month_counts_df.sort_values(by='month')

month_counts_df.to_csv('histogram_data.csv', index=False)
```

```python
result = final_df.groupby('season')['year_planted'].value_counts()
sorted_result = result.sort_index(ascending=[True, True])
sorted_result
```

```python
sorted_df = sorted_result.reset_index(name='count')
sorted_df.to_csv('sorted_trees.csv', index=False)
```
